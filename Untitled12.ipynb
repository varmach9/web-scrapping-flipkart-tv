# **Feature selection**

Finding inches of tv

```python
incharr=[]
for i in range(len(df)):
  print(df["des"][i])
  s=""
  for i in df["des"][i].lower():
    if i not in ")(":
      s+=i
  arr=s.split(" ")
  there=0
  for i in range(len(arr)):
    if i!=len(arr)-1 and arr[i+1] in ["inch","inches"]:
      incharr.append((arr[i]))
      there=1
  if not there:
    incharr.append(None)
print(len(df),len(incharr))
```

```python
df["inch"]=incharr
```

```python
df
```

price extraction

```python
import re
pricearr=[]
for j in range(len(df)):
  s=""
  for i in df["cost"][j]:
    if i not in ",₹%" :
      s+=i
    if i in "₹%" and s!="":
      s+=" "
  a=s.split(" ")
  pricearr.append(int(a[0]))
print(pricearr)
```

```python
df["cost"]=pricearr
```

extracting pixels,hdmi and usb

```python
pixelarr=[]
hdmiarr=[]
usbarr=[]
for i in range(len(df)):
  s=""
  for k in df["fea"][i]:
    if k!="X" and k!="x":
      s+=k
    else:
      s+=" "
  #print(s)
  arr=s.split(" ")
  arr=[ii for ii in arr if ii]
  there=0
  there2=0
  for j in range(len(arr)):

    #pixels
    if arr[j].lower()=="pi":
      if arr[j-1][-1]=="p":
        arr[j-1]=arr[j-1][:-1]
      if arr[j-1].isnumeric():
        pixelarr.append(int(arr[j-1]))
        there=1
    
    #hdmi
    if j>=1 and arr[j].lower()=="hdmi":
      hdmiarr.append(int(arr[j-1][-1]))
      there2=1
    
    #usb
    if j>=1 and len(arr[j])>=3 and arr[j][:3].lower()=="usb":
      if arr[j-1][-1].isdigit():
        usbarr.append(int(arr[j-1][-1]))
      else:
        usbarr.append(None)

  if not there:
    pixelarr.append(None)
  if not there2:
    hdmiarr.append(None)
print(len(usbarr))
```

```python
df["pixels"]=pixelarr
df["hdmi"]=hdmiarr
df["usb"]=usbarr
```
